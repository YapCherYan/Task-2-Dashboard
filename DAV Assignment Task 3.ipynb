```python
import pandas as pd
from dash import Dash, html, dcc, dash_table
import plotly.express as px
```

```python
df = pd.read_csv("DAV Assignment Task 3.csv")

print(df.head())
print(df.columns)
```

```python
line_fig = px.line(
    df,
    x="Year",
    y=["GDP Growth Rate", "Unemployment Rate"],
    title="GDP Growth Rate vs Unemployment Rate (2000–2024)",
    markers=True
)
```

```python
selected_years = df[(df["Year"] >= 2000) & (df["Year"] <= 2024)]

bar_fig = px.bar(
    selected_years,
    x="Year",
    y=["GDP Growth Rate", "Unemployment Rate"],
    barmode="group",
    title="Comparison of GDP Growth Rate and Unemployment Rate (2000-2024)"
)
```

```python
avg_gdp = df["GDP Growth Rate"].mean()
avg_unemp = df["Unemployment Rate"].mean()
latest_year = df.iloc[-1]

app = Dash(__name__)
```

```python
app.layout = html.Div(style={"fontFamily": "Arial", "padding": "20px"}, children=[

    html.H1(
        "Malaysia Economic Dashboard",
        style={"textAlign": "center", "color": "#2c3e50"}
    ),

    html.P(
        "Research Question: What is the relationship between unemployment rate and GDP growth rate in Malaysia, and what differences can be observed between 2000 and 2024?",
        style={"textAlign": "center"}
    ),

    # KPI Cards
    html.Div(style={"display": "flex", "justifyContent": "space-around", "marginTop": "20px"}, children=[

        html.Div([
            html.H4("Average GDP Growth"),
            html.H2(f"{avg_gdp:.2f}%")
        ], style={
            "backgroundColor": "#3498db",
            "color": "white",
            "padding": "20px",
            "borderRadius": "10px",
            "width": "30%",
            "textAlign": "center"
        }),

        html.Div([
            html.H4("Average Unemployment"),
            html.H2(f"{avg_unemp:.2f}%")
        ], style={
            "backgroundColor": "#e74c3c",
            "color": "white",
            "padding": "20px",
            "borderRadius": "10px",
            "width": "30%",
            "textAlign": "center"
        }),

        html.Div([
            html.H4("Latest Year"),
            html.H2(f"GDP: {latest_year['GDP Growth Rate']}% | Unemp: {latest_year['Unemployment Rate']}%")
        ], style={
            "backgroundColor": "#2ecc71",
            "color": "white",
            "padding": "20px",
            "borderRadius": "10px",
            "width": "30%",
            "textAlign": "center"
        }),

    ]),

    # Charts
    html.Div(style={"display": "flex", "marginTop": "40px"}, children=[

        html.Div([
            html.H3("Trend Analysis"),
            dcc.Graph(figure=line_fig)
        ], style={"width": "50%", "padding": "10px"}),

        html.Div([
            html.H3("Key Year Comparison"),
            dcc.Graph(figure=bar_fig)
        ], style={"width": "50%", "padding": "10px"}),

    ]),

    # Table
    html.H3("Dataset Overview", style={"marginTop": "30px"}),

    dash_table.DataTable(
        data=df.to_dict('records'),
        columns=[{"name": i, "id": i} for i in df.columns],
        page_size=10,
        style_table={'height': '300px', 'overflowY': 'auto'},
        style_cell={'textAlign': 'center', 'padding': '5px'},
        style_header={'backgroundColor': '#34495e', 'color': 'white'}
    ),

    # Insight
    html.Div(style={"marginTop": "30px"}, children=[
        html.H3("Key Insights"),
        html.P(
            "The results show a negative relationship between GDP growth Rate and unemployment Rate. "
            "During economic downturns such as 2009 and 2020, unemployment increased. "
            "During recovery periods, unemployment decreased as GDP improved."
        )
    ])

])

if __name__ == '__main__':
    app.run(jupyter_mode='external', port=8053)
```
